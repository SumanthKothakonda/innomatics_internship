```python
# Take input from user
sentence = input("Enter sentence: ")

# Split sentence into words
words = sentence.split()

# Convert list to set to get unique words
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

```python
numbers = [45, 22, 89, 10, 66]
print("List:",numbers)

#using max for highest number
print("Max:",max(numbers))

#using min for lowest number
print("Min:",min(numbers))
```

```python
prices = [450, 1200, 899, 1500, 300]
#empty list to store values greater than 1000
a=[]
for i in prices :
    #appending only values >1000
    if i>1000:
        a.append(i)
print("Products above 1000:",len(a))
```

```python
attendance = ["P", "P", "A", "P", "P"]
days=len(attendance)
p=0
for i in attendance:
    if i=="P":
        p+=1
print("Attendance Percentage :",(p/days)*100)
        
```

```python
text = "pythonp"
char_count = {}
for i in text:
    char_count[i] = char_count.get(i, 0) + 1
print(char_count)
```
