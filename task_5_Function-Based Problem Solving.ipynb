```python
#Problem 1: Social Media – Post Engagement Analyzer

def analyse_engagement(likes):
    total_likes=0
    for like in likes:
        total_likes+=like
    if total_likes>=1000:
        status="Viral Post"
    else:
        status="Normal Engagement"
    print("Total Likes = ",total_likes)
    print("Post Status = ",status)

likes = [100,200,52,185,398,159]
analyse_engagement(likes)
```

```python
#Problem 2: Healthcare – Medicine Stock Alert System

def medicine_stock(medicine):
    if medicine<10:
        Status = "Low Stock Alert"
    else:
        Status = "Stock Sufficient"
    print("Status = ",Status)
    
medicine=int(input("Medicine Stock :"))
medicine_stock(medicine)
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker

def rainfall_analyse(rainfall_data):
    total=0
    for data in rainfall_data:
        total+=data
    average = total/len(rainfall_data)
    if average>=required_level :
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print("Average Rainfall:",average)
    print("Rainfall Status:",status)

rainfall_data=[70,58,65,90,75]
required_level=70
rainfall_analyse(rainfall_data)
```

```python
#Problem 4: Social Media – Duplicate Account Detection

def acc_detection(usernames):
    total_users=len(usernames)
    usernames_set = set(usernames)
    unique_users = len(usernames_set)
    if (total_users == unique_users):
        status = "No"
    else :
        status = "Yes"
    print("Duplicate Accounts Found:",status)

usernames=['aaa','bbb','ccc','aaa','ddd','ccc']
acc_detection(usernames)
```

```python
#Problem 5: Healthcare – Appointment Eligibility Checker

def eligibility(patient_age):
    if patient_age>=18:
        eligibility = "Eligible"
    else :
        eligibility = "Not Eligible"
    print("Eligibility Status : ",eligibility)

patient_age=int(input("Patient Age :"))
eligibility(patient_age)
```

```python
#Problem 6: Agriculture – Premium Crop Price Filter

def crops_filter(crops):
    premium_crops=[]
    for crop in crops:
        if crop>2000:
            premium_crops.append(crop)
    print("Premium Crops",premium_crops)

crops=[1500,2500,3200,1800,1450]
crops_filter(crops)
```

```python
#Problem 7: System Monitoring – Application Health Checker

def health_check(error_count):
    if error_count==0:
        status="Healthy"
    elif error_count<=5:
        status="Minor Issues"
    else:
        status="Critical Issues"
    print("System Status :",status)
error_count=int(input("Error Count :"))
health_check(error_count)
```

```python
#Problem 8: Banking – Daily Transaction Limit Checker

def limit_checker(transaction_amount):
    if transaction_amount<=Daily_Limit:
        status = "Approved"
    else :
        status = "Rejected"
    print("Transaction Status : ",status)

transaction_amount=int(input("Transaction Amount :"))
Daily_Limit=50000
limit_checker(transaction_amount)
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System

def exam_eligibility(attendance_list):
    total_count=len(attendance_list)
    present_count=0
    for attendance in attendance_list:
        if attendance=='P':
            present_count+=1
    attendance_percentage= round(((present_count/total_count)*100),1)
    if attendance_percentage>=75:
        eligibility="Eligible"
    else:
        eligibility="Not Eligible"
    print("Attendance Percentage :",attendance_percentage)
    print("Exam Eligiblity : ",eligibility)
attendance_list=['P','P','P','P','P','F']
exam_eligibility(attendance_list)
```

```python
#Problem 10: Smart Electricity Bill Analyzer

def bill_analyzer(units_consumed):
    if units_consumed<=100:
        bill=units_consumed*3
    elif units_consumed<=200:
        bill = (100*3) + ((units_consumed-100)*5)
    else :
        bill = (100*3) + (100*5) + ((units_consumed-200)*7) 

    if bill<500:
        usage = "Low Usage"
    elif bill>=500 and bill<1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill , usage


units_consumed=int(input("Units Consumed :"))
bill,usage=bill_analyzer(units_consumed)
print("Electricity Bill :",bill)
print("Usage Level :",usage)
```
