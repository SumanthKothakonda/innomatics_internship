```python
#Problem Statement 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

#Find highest score
max_score = max(employees.values())

top_performers = []

for name in employees:
    if employees[name] == max_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {max_score})")
```

```python
#Problem Statement 2: Search Query Keyword Analysis

text = "Buy mobile phone buy phone online"

# Convert to lowercase
text = text.lower()

# Split words
words = text.split()

# Count frequency
count = {}

for word in words:
    count[word] = count.get(word, 0) + 1

result = {}

for word in count:
    if count[word] > 1:
        result[word] = count[word]

print(result)
```

```python
#Problem Statement 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:   
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Problem Statement 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domains = []

# Get only domain names
for email in emails:
    domains.append(email.split("@")[1])

total = len(domains)

for domain in set(domains):
    percent = domains.count(domain) * 100 // total
    print(domain + ":", str(percent) + "%")
```

```python
#Problem Statement 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

avg = sum(sales) / len(sales)
limit = avg + (avg * 0.3)

for i in range(len(sales)):
    if sales[i] > limit:
        print("Day", i + 1, ":", sales[i])
```

```python
#Problem Statement 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

for user in user_ids:
    count[user] = count.get(user, 0) + 1

# Print duplicates only
for user in count:
    if count[user] > 1:
        print(user, "→", count[user], "times")
```
