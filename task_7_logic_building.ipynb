```python
#Problem 1: Smart Light Controller

class SmartLight:

    def __init__(self,name):
        self.name=name
        self.status="OFF"
        
    def control(self,action):
        self.status=action
        print(self.name,"is",self.status)
    

name = input("Light Name :")
action = input("Action :")

light = SmartLight(name)

light.control(action.upper())
```

```python
#Problem 2: Employee ID Card System

class EmpId:

    def __init__(self,name,emp_id,dept):
        self.name=name
        self.emp_id=emp_id
        self.dept=dept

    def  id_card(self):
        print("-"*25)
        print("Employee ID Card")
        print("Name :",self.name)
        print("ID :",self.emp_id)
        print("Department :",self.dept)
    

name = input("Employee Name :")
emp_id = input("Employee ID :")
dept = input("Department :")

card = EmpId(name,emp_id,dept)

card.id_card()
```

```python
#Problem 3: Mobile Contact Record

class Mobile :
    def __init__(self,name,num):
        self.name=name
        self.num=num

    def contact_record(self):
        print("-"*25)
        print("Contact Saved")
        print("Name :",self.name)
        print("Phone :",self.num)

name = input("Contact Name :")
num = input("Phone Number :")

contact=Mobile(name,num)

contact.contact_record()
```

```python
#Problem 4: Product Price Tag Generator

class Price_Tag:
    
    def __init__(self,prod_name,price):
        self.prod_name = prod_name
        self.price = price
    def price_tag_generator(self):
        print("-"*25)
        print("Product :",self.prod_name)
        print("Price :","₹"+str(self.price))
        

prod_name=input("Product Name :")
price = int(input("Price :"))

tag = Price_Tag(prod_name , price)

tag.price_tag_generator()
```

```python
#Problem 5: Movie Rating Display System

class MovieRating:

    def __init__(self,name,rating):
        self.name=name
        self.rating=rating

    def MovieRatingDisplay(self):
        print("-"*25)
        print("Movie :",self.name)
        print("Rating :",str(self.rating)+"/5")

name = input("Movie Name :")
rating = float(input("Rating :"))

movie_rating = MovieRating(name,rating)

movie_rating.MovieRatingDisplay()
```

```python
#Problem 6: Delivery Address Manager

class DeliveryAddress:

    def __init__(self,name,address):
        self.name=name
        self.address=address

    def Delivery_Address_Manager(self):
        print("-"*25)
        print("Delivery Details")
        print("Customer :",self.name)
        print("Address :",self.address)

name=input("Customer Name :")
address=input("Address :")

del_add = DeliveryAddress(name,address)

del_add.Delivery_Address_Manager()
```
