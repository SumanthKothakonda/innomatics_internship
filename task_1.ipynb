```python
#1.User Login Check

name = str(input("Enter Username :"))
password = int(input("Enter Password :"))
if name == "admin" and password == 1234 :
    print("Login Successful")
else :
    print("Invalid Credentials")
```

```python
#2. Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]
Pass=0
Fail=0
for i in marks:
    if i>=50:
        Pass+=1
    else:
        Fail+=1
print("Total Number of Pass Students :",Pass)
print("Total Number of Fail Students :",Fail)
```

```python
#3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]
Names =[]
for i in names :
    a=i.lower()
    b=a.replace(" ","")
    Names.append(b)
print(Names)
```

```python
#4. Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
    length = len(i)
    print(i,":",length)
    if length > 10:
        print("Message longer than 10 characters")
```

```python
#5. Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs :
    if i=="ERROR":
        count+=1
print("Total ERROR Count :",count)
```
