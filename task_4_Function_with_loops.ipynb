```python
# Problem 1: Mobile Recharge Validation System

def recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    
    if amount not in valid_plans:
        print("Invalid plan. Available plans are: 199, 299, 399, 599.")
        return False
    
    return True

while True:
        amount = int(input("Enter recharge amount: ₹"))
        
        if recharge(amount):
            print("Recharge successful!")
            break
        else:
            print("Please try again.\n")
```

```python
#Problem 2: Inventory Reorder Alert System

def stock(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(product,": Reorder Alert (Stock =",quantity,")")
        else:
            print(product,": Stock OK (Stock =",quantity,")")

prod_stock = {
    "Rice": 29,
    "Wheat": 8,
    "Sugar": 18,
    "Oil": 14,
    "Salt": 35
}

stock(prod_stock)
```

```python
# Problem 3: Student Result Processing System

def student(marks):
    total = 0
    
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


student_marks = [67, 49, 70, 58, 42]

avg, result = student(student_marks)

print("Average Marks:",avg)
print("Result:",result)
```

```python
#Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if peak:
        fare = fare + (fare * 0.25)  
    
    return fare


while True:
    try:
        distance = float(input("Enter distance traveled (in km): "))
        peak = input("Is it peak hour? (yes/no): ").lower()
        
        if peak == "yes":
            peak = True
        elif peak == "no":
            peak = False
        else:
            print("Invalid input for peak hour. Please enter yes or no.\n")
            continue
        
        total_fare = calculate_fare(distance, peak)
        
        print("Total Fare:",total_fare)
        break
    
    except ValueError:
        print("Please enter a valid distance.\n")
```

```python
#Problem 5: Employee Attendance Eligibility Checker

def eligibility(attendance_list):
    present_days = 0
    
    for status in attendance_list:
        if status == "P":
            present_days += 1
    
    total_days = len(attendance_list)
    percentage = (present_days / total_days) * 100
    
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


attendance = ["P", "A", "P", "P", "A", "P", "P", "P","A"]

percent, result = eligibility(attendance)

print("Attendance Percentage:",round(percent,2))
print("Status:",result)
```

```python
#Problem 6: Password Strength Checker

def check_password_strength(password):
    has_digit = False
    has_special = False

    if len(password) < 8:
        return False, "Minimum 8 characters required."

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return True, "Strong Password "
    else:
        return False, "Must contain at least one digit and one special character (@#$)."


while True:
    pwd = input("Enter password: ")
    is_valid, message = check_password_strength(pwd)

    if is_valid:
        print(message)
        break
    else:
        print("Weak Password", message)
        print("Please try again.\n")
```
